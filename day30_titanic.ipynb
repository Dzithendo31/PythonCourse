## Pre-processiong
1. Feature Selection
2. Clean them null vals


```python
feature_df = df[['Age','Pclass','Fare']]
target_df = df[['Survived']]
feature_df.isnull().sum()
```

```python
#Imput the Missing vlaue with mean
#Using a Skikit
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="mean")
feature_df_imputed = imputer.fit_transform(feature_df)
#Convert it into a List/Array instead of a DF
feature_df_imputed
```

# 3. Standardizing
In machine learning, "standardizing" refers to a preprocessing step where the features of a dataset are transformed in a way that they have zero mean and unit variance. 

Standardizing is important because it can help improve the performance and convergence speed of certain machine learning algorithms, particularly those that are sensitive to the scale of the input features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#Stnadrise your value based on your means.
feature_scaled = scaler.fit_transform(feature_df_imputed)
feature_scaled
```

```python
#Check if the is a chnage
imputed_df = pd.DataFrame(feature_df_imputed)
imputed_df.describe()
```

# Modeling
20% using for test data, and 80% used for 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature_scaled,target_df,test_size=0.02,random_state=42)
```

# Using the Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

Model Ready to predict

```python
#Prediction process
predict = model.predict(X_test)
predict
```

```python
y_test
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,predict)
print(f"Model accuracy:{accuracy:.2%}")
```
