```python
import pandas as pd
file = "loan.csv"

df = pd.read_csv(file)
df.head()
```

Credit_History

##### OutPut
Loan_Status

```python
features = df[['Credit_History','Loan_Amount_Term','ApplicantIncome','CoapplicantIncome']]
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})
target_df = df[['Loan_Status']]
target_df
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="median")
feature_df_imputed = imputer.fit_transform(features)
#Convert it into a List/Array instead of a DF
feature_df_imputed
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#Stnadrise your value based on your means.
feature_scaled = scaler.fit_transform(feature_df_imputed)
feature_scaled
```

```python
imputed_df = pd.DataFrame(feature_df_imputed)
imputed_df.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature_scaled,target_df,test_size=0.5,random_state=42)
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
predict
```

```python

accuracy = accuracy_score(y_test,predict)
print(f"Model accuracy:{accuracy:.2%}")
```
