```python
tips_df = pd.read_csv("./tips.csv")
tips_df
```

```python
tipPp = tips_df['tip']/tips_df['size']
tips_df['tipPp'] = tipPp
tips_df.tail(10)
```

```python
#How much % peopl are tipping
tipPercent = (tips_df['tip']/tips_df['total_bill'])*100
tips_df['tipPercent '] = round(tipPercent,2) 
tips_df.tail(10)
```

## Sorting and Stuff

```python
tips_df.sort_values(by='total_bill',ascending=False)
```
