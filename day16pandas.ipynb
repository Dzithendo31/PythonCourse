### Group by

Average total bill on any day that you take
- So like Average for Sunday or Wed
```sql
Select average(total_bill)
from tips_df
group by day
```

````python
tips_df.groupby('day')['total_bill'].mean()
````

Who is leaving more tips between smokers

````python
tips_df.groupby('smoker')['tip'].mean()
````

````python
tips_df.groupby('smoker')['tip'].sum()
````

#### Task 03

````python
tips_df.groupby('day').aggregate({'total_bill':'sum','tip':'mean'})
````

# One More level of drilling

````python
#The order will matter in terms of the stuff that happend here
tips_df.groupby(['smoker','day'])['tip'].sum()
````

#### Task who earns more tip each day

````python
#The order will matter in terms of the stuff that happend here
tips_df.groupby(['day','sex'])['tip'].mean().unstack()
````

##Task
max min and who earns more in each day


````python
toCSV = tips_df.groupby(['day','sex']).aggregate({'tip':['max','min']}).unstack()
toCSV.to_csv("./group.csv")
````

# Bar Graph
Average Total bill by each day

````python
x_days = tips_df['day'].unique()

y_avg = tips_df.groupby('day')['total_bill'].mean()

plt.bar(x_days,y_avg,color="pink",alpha=0.5)
plt.xlabel("Days")
plt.ylabel("Average Sales in R")
plt.title("Average Sales In the Week")
````

````python
y_avg = tips_df.groupby('day')['total_bill'].mean()
x_days = y_avg.index

y_avg.plot(kind="bar", xlabel="Days", ylabel="Average sales in R", title="Average sales in the week", color="crimson",  alpha=0.2)
````

````python
y_avg = tips_df.groupby('day')['total_bill'].mean()
x_days = y_avg.index

y_avg.plot(kind="pie", autopct='%1.1f%%', title="Average sales in the week")
````
